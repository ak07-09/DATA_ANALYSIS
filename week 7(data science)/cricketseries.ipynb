### balls: 60 70 30 95 110 65 75 100 45 35 85 100 120 55 70 80 85 60 40 50 80 105 115 55 70 90 55 


```python
#1 what is the total number of runs scored in the series

import numpy as np
runs = np.array([
    54, 78, 23, 89, 101, 56, 45, 67, 90, 33,
    23, 76, 85, 95, 41, 65, 75, 80, 45, 29,
    34, 78, 92, 101, 50, 60, 71, 66, 85, 47
])

balls = np.array([
    60, 70, 30, 95, 110, 65, 50, 75, 100, 45,
    35, 85, 100, 120, 55, 70, 80, 85, 60, 40,
    50, 80, 105, 115, 55, 70, 80, 75, 90, 55
])


print(runs.sum())   
```

```python
#2 What is the average number of runs scored per match

print(runs.mean())
```

```python
# What is the average strike rate of the player

print(runs/balls*100)
```

```python
# What is the shape of the runs array

print(runs.shape)    
```

```python
# What is the number of runs scored by the 5th player 

print(runs[4])
```

```python
# what is the toatal number of runs scored in the last five matches?


print(runs[-5:])
```

```python
# Total runs scored by players who faced more than 80 balls

print(runs[balls>80].sum()) 
```

```python
# identify the indices where players scored more than 80 runs

print(np.where(runs>80))
```

```python
#  Extract the runs for the 1st, 3rd, and 5th players from the runs array.
 
print(runs[[0,2,4]])
```

```python
#  Find the runs scored on the 1st and 2nd matches and 4th and 5th matches.

print(runs[[0,1,3,4]]) 
```
