## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load dataset


```python
df = pd.read_csv('WineQT.csv')
df.head()
```

## Explore data 


```python
print("Shape of dataset:", df.shape)
print("\nInfo about dataset:\n")
df.info()
print("\nMissing values in each column:\n", df.isnull().sum())
print("\nSummary statistics of dataset:\n", df.describe(include='all'))
```

## Insights

1. Dataset contains various physicochemical properties like acidity, alcohol, etc.

2. No missing data detected.

3.  Target variable 'quality' is a score (categorical-like numerical).

## 6. Data Cleaning

```python
df.fillna(df.median(), inplace=True)
```

## 7. Basic Statistical Analysis

```python
print("\nMean values for numerical features:\n", df.mean())
print("\nMedian values for numerical features:\n", df.median())
print("\nDistribution counts for 'quality':\n", df['quality'].value_counts())
```

## 8. Visualization: Wine Quality Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=df)
plt.title('Wine Quality Distribution')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.show()
```

## Insight:

1. Quality scores range from around 3 to 8, with 5 and 6 being most common.

2. Distribution is slightly skewed, more wines have mid-range quality.

## 9. Visualization: Feature Correlation Heatmap



```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Features')
plt.show()
```

## Insight:
1. Alcohol content positively correlates with quality.

2. Sulphates and citric acid also show positive correlation with quality.

3. Density negatively correlates with quality.

4. High correlations exist among some features (e.g., fixed acidity and citric acid), indicating multicollinearity.


## 10. Visualization: Alcohol Content vs Quality Boxplot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='alcohol', data=df)
plt.title('Alcohol Content vs Wine Quality')
plt.xlabel('Quality Score')
plt.ylabel('Alcohol Percentage')
plt.show()
```

## Insight:

1. Higher alcohol content tends to be associated with higher quality ratings.

2. Boxplots show clear increasing trend of median alcohol with quality.

# 11. Final Insights

1. The chemical composition clearly influences wine quality.

2. Alcohol content is a key positive factor.

3. The data distribution and correlation patterns provide important guidance for predictive modeling.
