## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv('Iris.csv')
df.head()
```

## Data Exploration

```python
print("Dataset shape:", df.shape)
print("\nInformation:")
df.info()
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nDescriptive statistics:")
print(df.describe(include='all'))
```

## Insights

1. Dataset has four features: sepal length, sepal width, petal length, petal width.

2. Target variable: species with three classes (setosa, versicolor, virginica).

3. No missing data.

## Distribution of Features by Species

```python
plt.figure(figsize=(15,10))

feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, feature in enumerate(feature_names):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=df, x=feature, hue='Species', kde=True, element='step', stat='density')
    plt.title(f'Distribution of {feature} by Species')

plt.tight_layout()
plt.show()
```

## Insight:

1. Petal length and petal width distributions clearly separate Iris-setosa from others.

2. Sepal features show overlapping distributions but still differ across species.

## Pairplot to Visualize Pairwise Relationships

```python
sns.pairplot(df, hue='Species', corner=True)
plt.suptitle('Pairwise Feature Relationships', y=1.02)
plt.show()
```

## Insight:

1. Pairplot reveals distinct clustering of species especially around petal length and width.

2. Strong evidence that these features help in classification.

## Correlation Heatmap

```python
plt.figure(figsize=(8,6))
correlation = df.loc[:, feature_names].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

## Insight:

1. Petal length and petal width are highly positively correlated.


2. Sepal width negatively correlates with petal length/width somewhat.

## Boxplots of Features by Species

```python

plt.figure(figsize=(12,8))

for i, feature in enumerate(feature_names):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} by Species')

plt.tight_layout()
plt.show()
```

## Insight:

1. Boxplots emphasize clear separation in median and spread for petal dimensions between species.

2. Sepal width shows more overlap but still varies.


# Final Insights

1. The Iris dataset features exhibit distinctive patterns across species.


2. Petal measurements are the strongest discriminators.

3. This foundational analysis sets the stage for predictive modeling tasks.
