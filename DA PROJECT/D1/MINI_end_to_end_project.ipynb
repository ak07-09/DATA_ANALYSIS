## 2. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 3. Load Dataset

```python
df = pd.read_csv('exams.csv')
df.head()
```

## 4. Data Overview

```python
print("Dataset shape:", df.shape)
print("\nDataset info:\n")
df.info()
print("\nMissing values:\n", df.isnull().sum())
print("\nDescriptive stats:\n", df.describe(include='all'))
```

## Insights

1. Dataset contains scores in math, reading, writing.

2. Also contains categorical info: gender, race/ethnicity, parental level of education, lunch, test preparation course.

3. No missing data found


## 6. Data Cleaning

#### No missing values found, no extra cleaning required at this point


## 8. Visualize Exam Score Distributions

```python
plt.figure(figsize=(10,6))
sns.histplot(df['math score'], kde=True, color='blue', label='Math Score')
sns.histplot(df['reading score'], kde=True, color='green', label='Reading Score')
sns.histplot(df['writing score'], kde=True, color='red', label='Writing Score')
plt.legend()
plt.title('Distribution of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

## Insight:
1. Scores roughly normally distributed.

2. Reading and writing scores show similar distributions.

3. Math score distribution shifts slightly left compared to reading/writing.

## 9. Visualize Scores by Gender Boxplots

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score Distribution by Gender')
plt.show()
```

## Insight:
1. Females and males have similar median math scores, with slight variations in spread.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Score Distribution by Gender')
plt.show()
```

## Insight:
1. Females tend to have slightly higher median reading scores.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Score Distribution by Gender')
plt.show()
```

## Insight:
1. Females appear to outperform males in writing scores.


## 10. Relationship Between Parental Level of Education and Average Score

```python
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

plt.figure(figsize=(14,7))
sns.barplot(x='parental level of education', y='average score', data=df, order=df['parental level of education'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Average Exam Score by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.show()
```

## Insight:

1. Students with parents having higher education levels tend to have better average scores.

2. Clear upward trend with education level of the parents.


## 11. Impact of Test Preparation Course on Average Scores

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='test preparation course', y='average score', data=df)
plt.title('Test Preparation Course vs Average Score')
plt.xlabel('Test Preparation Course Completed')
plt.ylabel('Average Score')
plt.show()
```

## Insight:

1. Students completing test preparation courses tend to score higher on average.

2. Indicates preparation course has positive impact on exam performance.

# Final Insights

1. Scores vary by gender, parental education, and test preparation completion.

2. Reading and writing scores are generally higher than math scores.

3. Preparation courses and parental education have positive influence on outcomes.
